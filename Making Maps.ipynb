```python
#Creating a Simple Map
import folium

map = folium.Map(location=(28, 77), control_scale=True, zoom_start=8)  #28,77 delhi

popup_text = f"<h1 style='color:red;'>Heyy, you clicked me!</h1>"

# Adding a Marker
folium.Marker(
    location=[28, 77],
    tooltip="Click Me",
    popup=popup_text
).add_to(map)

# Adding a Polyline
coordinates = [[28.10, 77.10], [29.10, 78.10]]


folium.PolyLine(
    locations=coordinates,
color='red',
tooltip="My awesome tooltip"
).add_to(map)

# Adding a polygon
locations = [
    [28.45, 76.47],
    [28.53, 77.13],
    [28.31, 77.26],
    [28.24, 76.58],
]

folium.Polygon(
    locations=locations,
    fill=True,
    fill_color='transparent',
    fill_opacity=1,
    no_clip=False
).add_to(map)

map
```
