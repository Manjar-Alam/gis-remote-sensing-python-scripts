```python
# Import necessary libraries
import folium  # For creating maps
from folium.plugins import MarkerCluster  # For clustering points
import numpy as np  # For generating random data

# Define the number of random points
size = 500

# Generate random latitude and longitude values within specified ranges
lats = np.random.randint(28, 38, size).tolist()  # Latitude range: 28 to 38
longs = np.random.randint(77, 87, size).tolist()  # Longitude range: 77 to 87

# Create a base map with an arbitrary center (latitude: 6, longitude: -1)
map = folium.Map(location=(6, -1))

# Automatically adjust the map's view to fit the generated points
map.fit_bounds([
    [float(np.min(lats)), float(np.min(longs))],  # Southwest corner of the bounding box
    [float(np.max(lats)), float(np.max(longs))]   # Northeast corner of the bounding box
])

# Create a cluster group to organize points on the map
points_cluster = MarkerCluster().add_to(map)

# Plot the points with alternating marker styles
for x, y in zip(lats, longs):
    # Use green camera icon for even-indexed points
    if lats.index(x) % 2 == 0:
        folium.Marker(
            location=(x, y),
            icon=folium.Icon(color='green', icon="camera")  # Green icon with a camera symbol
        ).add_to(points_cluster)
    # Use blue tag icon for odd-indexed points
    else:
        folium.Marker(
            location=(x, y),
            icon=folium.Icon(color='blue', icon="tag")  # Blue icon with a tag symbol
        ).add_to(points_cluster)

# Display the map (works in Jupyter Notebook or interactive environments)
map
```
